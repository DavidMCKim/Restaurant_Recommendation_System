```python
def crawl_review(localname):
    driver = webdriver.Chrome('/Users/mckimair/chromedriver')
    for i in range(len(f'{localname}_matzip_info_df')):

        review_url = 'https://pcmap.place.naver.com/accommodation/{}/review/visitor?'.format(f'{localname}_matzip_info_df'['Num'][i])
        try:

            # 캠핑장 리뷰 개별 url 접속
            driver.get(review_url)
            time.sleep(2) 

            while True:
                try:
                    # END 키를 눌러서 페이지 제일 하단으로 이동
                    driver.find_element_by_tag_name('body').send_keys(Keys.END)
                    time.sleep(2)
                    # 더보기 버튼 클릭
                    driver.find_element_by_xpath('//*[@id="app-root"]/div/div/div[2]/div[5]/div[4]/div[3]/div[2]/a').click()
                    time.sleep(2)
                    # END 키를 눌러서 페이지 제일 하단으로 이동
                    driver.find_element_by_tag_name('body').send_keys(Keys.END)
                    time.sleep(1)

                except NoSuchElementException:
                    break

            # 파싱
            html = driver.page_source
            soup = BeautifulSoup(html, 'lxml')
            time.sleep(1)

            # 캠핑장 구분
            matzip_name = f'{localname}_matzip_info_df'['Name'][i]

            try:
                all_review = soup.find_all('li', attrs = {'class':'_3FaRE'})
                review_num = len(all_review) # 특정 캠핑의 리뷰 총 개수


                for review in range(review_num):
                    ## 캠핑장 고유 넘버 
                    matzip_num_list.append(f'{localname}_matzip_info_df'["Num"][i])

                    ## 캠핑장주소
                    matzip_addr_list.append(f'{localname}_matzip_info_df'["Address"][i])

                    ## 캠핑장 이름
                    matzip_name_list.append(matzip_name)

                    ## user ID
                    user = soup.find_all('div',{'class':'_16RxQ'})
                    user_id_list.append(user[review].text)

                    ## rating(사용자가 이 캠핑장에 준 별점)
                    try:
                        if len(all_review[review].find_all('span', attrs = {'class':'_1fvo3'})) == 6:
                            rating = all_review[review].find_all('span', attrs = {'class':'_1fvo3'})[3].em.text
                        else:
                            rating = 'None'
                    except:
                        rating = 'error'
                    rating_list.append(rating)

                    ## date
                    try:
                        date = all_review[review].find('div', attrs = {'class':'_3-LAD'}).time.text
                    except:
                        date = 'error'
                    date_list.append(date)                

                    ## count
                    try:
                        count = all_review[review].find('div', attrs = {'class':'_3-LAD'}).text[-10:]
                    except:
                        count = 'error'
                    count_list.append(count)    

                    ## review 내용
                    try: 
                        review_content = all_review[review].find('span', attrs = {'class':'WoYOw'}).text
                    except: # 리뷰가 없다면
                        review_content = "error"
                    review_list.append(review_content)                    


            except Exception as total_review_err:
                none_review = "네이버 리뷰 0개"


        except Exception as err:
            error_list = '{}행 에러 : {}'.format(i,err)
            error.append(error_list)
            rating_list.append(' ')
            avg_rating_list.append(' ')
            review_list.append(' ')

    driver.quit()    
    
    review_data = {'Num':matzip_num_list,'Address':matzip_addr_list,'Date':date_list, 'Name':matzip_name_list, 'User_ID':user_id_list, 'Rating':rating_list, 
               'Review':review_list}
    review_df = pd.DataFrame(review_data)
    return 
```
